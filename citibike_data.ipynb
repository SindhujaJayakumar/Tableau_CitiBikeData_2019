```python

import pandas as pd
import numpy as np
```

```python
# Read in all 2019 rider data
january_2019 = pd.read_csv('citi_bike_data/JC-201901-citibike-tripdata.csv')
february_2019 = pd.read_csv('citi_bike_data/JC-201902-citibike-tripdata.csv')
march_2019 = pd.read_csv('citi_bike_data/JC-201903-citibike-tripdata.csv')
april_2019 = pd.read_csv('citi_bike_data/JC-201904-citibike-tripdata.csv')
may_2019 = pd.read_csv('citi_bike_data/JC-201905-citibike-tripdata.csv')
june_2019 = pd.read_csv('citi_bike_data/JC-201906-citibike-tripdata.csv')
july_2019 = pd.read_csv('citi_bike_data/JC-201907-citibike-tripdata.csv')
august_2019 = pd.read_csv('citi_bike_data/JC-201908-citibike-tripdata.csv')
septemper_2019 = pd.read_csv('citi_bike_data/JC-201909-citibike-tripdata.csv')

```

```python
# Merge the 2019 monthly data and sort by trip_id
citibike_2019 = pd.concat([january_2019, february_2019, march_2019, april_2019, may_2019,
                       june_2019, july_2019, august_2019,septemper_2019])
citibike_2019 = citibike_2019.sort_values(by='starttime')
```
